```python
df = db.query("""
            select 
                --order_id,
                p.promo_name,
                sum(order_quantity) as quantity
            
            from read_csv_auto('../Dataset/orders (2).csv') o
            left join read_csv_auto('../Dataset/promotions (1).csv') p on o.order_date between p.start_date and p.end_date
            where p.promo_id is not null
            group by 1
            order by 2 desc
              
              
              
              """).to_df()

df
```
